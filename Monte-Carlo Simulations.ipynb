## 1. Calculating Pi

I first did this calculation as part of an assignment and really liked how simple the idea was to calculate $\pi$. 

The method involves the use of a square, raindrops and the known area of a circle.

$$A_o = \pi r^2$$
$$A_{\square} = r^2$$

This method actually uses a quadrant of a unit circle drawn on a square, rather than a full circle, and counts the number of raindrops that fall in the quarter circle compared to the square. 

$$\frac{\frac{A_o}{4}}{A_{\square}} = \frac{\pi}{4}$$

```python
# Create 2 Arrays of random x and y values
x = np.array(np.random.rand(100))
y = np.array(np.random.rand(100))
```

```python
# plot each (x,y)
fig, ax = plt.subplots(figsize=(7,7))
plt.plot(x,y,'bo')
plt.title('Visualisation of raindrops on the slab')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0,1)
plt.ylim(0,1)

# Draw the quadrant
circle = np.linspace(0,1,100)
plt.plot(np.cos((np.pi/2)*circle),np.sin((np.pi/2)*circle),'g-');
```

The next step is to counts the number of raindrops that occur within the circle out of the N total raindrops within the square.

```python
def Pi(N):
    x = np.array(np.random.rand(N))
    y = np.array(np.random.rand(N))
    i = 0
    counts = 0
    # Repeats for each of the 100 raindrops
    while i < N:
        # If any x,y pair satisfy this identity, a count is taken
        if x[i]**2+y[i]**2 <= 1:
            counts+=1
            i+=1
        else:
            i+=1
    return print('\u03C0 = ', 4*(counts/N))

Pi(1000)
Pi(10000)
Pi(100000)
Pi(1000000)
```

 ### 2.1 Drunken Sailor 
 
 This is a simplified example of the diffusion of particles that I computed to obtain a better understanding of how to calculate the distance of a particle. I am including this because I enjoyed the idea of comparing a particle moving through a gas to a drunk guy walking through a bar and forgetting which way was walking after each step.
 
 In this example, the assumed length of a stride for the drunken sailor is going to be 0.5m which is smaller than the average step due to a lack of coordination, I am comparing the step of a person my height so it has errors on the distance.

The following code is the mean distance travelled per step after a number of steps from his initial position.

```python
def Sailor(step,path):
    # array for the mean travel by sailor per step 
    m = np.zeros((2,step))
    # for each number of steps
    for i in range(0,step):
        # Array for each path displacement 
        paths = np.zeros(path)
        for j in range(0,path):
            position = np.zeros(2)
            # Assigns a random direction of the sailor to step
            for k in range(0,i+1):
                direction = np.random.uniform(0,1)
                if direction <= 0.25:
                    position[1]+=0.5
                elif direction > 0.25 and direction <= 0.5:
                    position[0]+=0.5
                elif direction > 0.5 and direction <= 0.75:
                    position[1]-=0.5
                elif direction > 0.75:
                    position[0]-=0.5
            paths[j] = np.sqrt(np.dot(position,position))
        m[0,i] = np.sum(paths)/(path)   
        m[1,i] = i+1
    return m
```

```python
x1 = Sailor(10,10000)
x2 = Sailor(10,100)
print(x1)
print(x2)
```

```python
plt.figure(figsize=(8,7))

plt.title('Mean step distance per n steps')
plt.plot(x1[1],x1[0],'bo-',label='No. paths = 10000')

plt.plot(x2[1],x2[0],'ro-',label='No. paths = 100')

plt.ylabel('Mean distance travelled (m)')
plt.xlabel('Number of steps taken')
plt.xlim(0,10)
plt.legend();
```

### 2,2 Diffusion in Air
Calculation of the average distance travelled in 1 second for 100 different particles with the following velocity and mean free path length:
$$ v = 500ms^{-1} $$
$$\lambda = 1m$$

```python
def Travel(velocity,length,time,path):
    # The average number of collisions
    N = int((velocity*time)/length)
    # Make array to assign a distance for each path
    paths = np.zeros(path)
    for i in range(0,path):
        initial = np.zeros(2)
        for j in range(1,N+1):
            x = np.random.uniform(0,2*np.pi)
            initial +=np.array([np.cos(x),np.sin(x)])
        paths[i] = np.sqrt(np.dot(initial,initial))
    distance = np.sum(paths)/path
    deviation = np.sqrt(np.dot((paths-distance),(paths-distance))/(path-1))
    return [distance,'\u00B1', deviation]
print(Travel(500,1,1,100))
print(Travel(500,1,1,1000))
```

Plot of histograms indicating the number of paths yielding displacements between 0 and 1 m, between 1 and 2 m, and so on, using 100, 200, 300, 400 and 500 different paths. 

```python
def Displace(velocity,length,time,n):
    N = int((velocity*time)/length)
    # Make array for assigning the distance of each path
    paths = np.zeros(n)
    for i in range(0,n):
        initial = np.zeros(2)
        for j in range(1,N+1):
            x = np.random.uniform(0,2*np.pi)
            initial+=np.array([np.cos(x),np.sin(x)])
        paths[i] = np.sqrt(np.dot(initial,initial))
    return paths

data1 = Displace(500,1,1,100)
data2 = Displace(500,1,1,200)
data3 = Displace(500,1,1,300)
data4 = Displace(500,1,1,400)
data5 = Displace(500,1,1,500)
```

```python
fig, (ax1,ax2,ax3,ax4,ax5) = plt.subplots(5,figsize=(10,40))

bins = np.arange(0,70,1)
ax1.set_title('Histogram of Displacement n=100')
ax1.hist(data1,bins)
ax1.set_xlabel('Displacement (m)')

ax2.set_title('Histogram of Displacement n=200')
ax2.hist(data2,bins)
ax2.set_xlabel('Displacement (m)')

ax3.set_title('Histogram of Displacement n=300')
ax3.hist(data3,bins)
ax3.set_xlabel('Displacement (m)')

ax4.set_title('Histogram of Displacement n=400')
ax4.hist(data4,bins)
ax4.set_xlabel('Displacement (m)')

ax5.set_title('Histogram of Displacement n=500')
ax5.hist(data5,bins)
ax5.set_xlabel('Displacement (m)');
```

Initially the histogram is a distribution skewed to the right. As the number of paths for the mean is increased the distribution becomes more evenly distributed and it becomes less skewed.

The next set of code monitors the distance of the particle after every collision and averages over a sufficiently large number of molecular paths to yield valid results. 
The average net displacement is then plotted as a function of the number of collisions.

```python
# collision is maximum number of collisions
# path is number of paths 
def Steptravel(collision,path):
    # array for the mean travel after N collisions 
    m = np.zeros((3,collision))
    # for each number of collisions
    for i in range(0,collision):
        paths = np.zeros(path)
        for j in range(0,path):
            position = np.zeros(2)
            for k in range(1,i+2):
                x = np.random.uniform(0,2*np.pi)
                position+=np.array([np.cos(x),np.sin(x)])
            paths[j] = np.sqrt(np.dot(position,position))
        m[0,i] = np.sum(paths)/(path)   
        m[1,i] = i+1
        m[2,i] = np.sqrt(np.dot((paths-m[0,i]),(paths-m[0,i]))/(path-1))
    return m 
x3 = Steptravel(500,1000) 
```

```python
plt.figure(figsize=(10,7))

plt.title('Mean distance travelled between collisions')
plt.xlabel('Number of collisions')
plt.ylabel('Mean distance travelled (m)')
plt.plot(x3[1],x3[0],'b-');
```

By producing a log-log plot of this data, it can then see that there must be a power dependence relation describing the distance travelled by a particle.

$$\frac{d}{\lambda} = N^\alpha$$
$$log\left(\frac{d}{\lambda}\right) = \alpha log(N)$$


```python
plt.figure(figsize=(12,7))
def linear(x,m,c):
    y = m*x+c
    return y

def linerror(x,dm2,dc2,dmc2):
    err = np.sqrt((x**2)*(dm2)+(dc2)+2*x*(dmc2))
    return err

def logerr(x,dx):
    err = dx/x
    return err

popt, pcov = curve_fit(linear,np.log(x3[1]),np.log(x3[0]), sigma=logerr(x3[0],x3[2]))

print('Power:',f'{popt[0]:.3}','\u00B1',f'{np.sqrt(pcov[0,0]):.1}')
print('Intercept:',f'{popt[1]:.3}','\u00B1',f'{np.sqrt(pcov[1,1]):.3}')
fit = linear(np.log(x3[1]),popt[0],popt[1])

yerr = linerror(np.log(x3[1]),pcov[0,0],pcov[1,1],pcov[1,0])

plt.title('log-log plot of data')
plt.plot(np.log(x3[1]),np.log(x3[0]),'y.',label='Data')
plt.plot(np.log(x3[1]),fit,'r-',label='Linear fit to data')
plt.plot(np.log(x3[1]),fit+yerr,'g--',label='Bounds of fit')
plt.plot(np.log(x3[1]),fit-yerr,'g--')
plt.ylabel('logarithm of mean distance travelled')
plt.xlabel('logarithm of number of collisions')
plt.xlim(0,6.5)
plt.legend();
```
