### Thermal energy Distribution

##### - Boltzmann Distribution 
 
    This is also known as the Maxwell-Boltzmann distribution and is very important in thermodynamics and is given by:

$$\bar{n} = e^{-\frac{\epsilon-\mu}{k_BT}}$$

##### - Bose-Einstein Distribution

    The average number of particles in a state with energy epsilon is:
$$\bar{n} = \frac{1}{e^{\frac{\epsilon-\mu}{k_BT}}-1}$$

    This equation arises from the fact that any number of bosons can exist in a state at a given energy.
    

##### - Fermi-Dirac Distribution

    The average number of fermions in a state with energy \epsilon is:
    
$$\bar{n} = \frac{1}{e^{\frac{\epsilon-\mu}{k_BT}}+1}$$
    
    This equation arises from the Pauli Exclusion Principle. The Pauli Exclusion Principle ensures that a state at a given energy either has 1 or no fermions as no 2 fermions can have the same energy and quantum state. 
    
    In the following defined formulae, E is unitless and represents the variance of particle energy from the mean particle energy for a given temperature.  
    
$$E = \frac{\epsilon-\mu}{k_BT}$$

```python
def Boltzmann(E):
    return np.exp(-E)

def Bose(E):
    return 1/(np.exp(E)-1)

def Fermi(E):
    return 1/(1+np.exp(E))
```

```python
y = np.linspace(0,3,200)
z = np.linspace(0.025,3,200)
B = Boltzmann(y)
BE = Bose(z)
FD = Fermi(y)
plt.title('Thermal State Distributions',fontsize=16)
plt.plot(y,B,'r-',label='Boltzmann Distribution')
plt.plot(y,BE,'g-',label='Bose-Einstein Distribution')
plt.plot(y,FD,'b-',label='Fermi-Dirac Distribution')
plt.ylim(0,6)
plt.ylabel('Number of particles in each state',fontsize=12)
plt.xlim(0,3)
plt.xlabel('(E-\u03BC)/kT',fontsize=12)
plt.legend()
plt.show();
```

### Blackbody Spectrum
    The Blackbody Spectrum is the distribution of energy emitted by a blackbody due to the temperature of the blackbody. This is a very important distribution to know when looking at the spectra of stars because it has some nice properties that help in the classification of stars.

##### - Rayleigh-Jeans Law
    This was the first theoretical formula determined for the blackbody spectrum. 

$$I(\lambda,T) = \frac{2\pi ckT}{\lambda^4}$$

##### - Planck's Law
    This formula was initially found by the quantisation of energy emitted. Although it wasn't known why at the time, this quantisation of energy is due to electrons only existing in constant shells or orbits around the nucleus.

$$I(\lambda,T) = \frac{2\pi hc^2}{{\lambda}^5(e^{(hc/\lambda kT)}-1)}$$

```python
def Planck(l):
    return 1/((l**5)*(np.exp(1/l)-1))

def RayJeans(l):
    return l**(-4)
```

```python
a = np.linspace(0.05,1,100)
b = np.linspace(0.3,1,100)
RJ = RayJeans(b)
Pl = Planck(a)

plt.title('Blackbody Spectrum',fontsize=16)
plt.plot(b,RJ,'r-',label='Rayleigh-Jeans Law')
plt.plot(a,Pl,'g-',label='Plancks Law')
plt.xlabel('Wavelength \u03BB (m)',fontsize=12)
plt.xlim(0,1)
plt.ylabel('Flux',fontsize=12)
plt.legend()
plt.show();
```
