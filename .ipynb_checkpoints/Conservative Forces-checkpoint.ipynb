### Conservative Forces

A mass acted on by a conservative force will have energy conservating. 

This notebook will evaluate the following force equation with its respective potential energy formula for this force acting on  a spring:
$$F(x) = -\frac{dU}{dx} = \kappa x-4q x^3$$
$$U(x) = -\frac{\kappa}{2}x^2+qx^3$$

This force and potential are evaluated for an oscillitary motion of a particle due to this force.

```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt;
```

From Newton's 2nd Law of motion:

    "The rate of change of momentum applied on an object is directly proportional to the force applied."
    
Assuming that no mass is lost (i.e. constant mass), this results in the following ordinary differential equation:
$$\frac{dv}{dt} = \frac{F}{m} $$
By definition of the velocity, a second ordinary differential equation can be found:
$$\frac{dx}{dt} = v $$



This can be written in matrix form:
$$\frac{d}{dt} \begin{pmatrix} v \\ x \\ \end{pmatrix} = \begin{pmatrix} \frac{F}{m} \\ v  \end{pmatrix} $$



#### Euler-Cromer
To solve this, the Euler-Cromer method wil be utilised to solve this system of ODEs.

The Euler-Cromer method is a loop with the following steps:
    
    1. Calculate the velocity, v, after a small change in time using the previous position:
$$ v_{i+1} = v_i+\frac{F(x_{i})}{m}\Delta t $$
    2. Calculate the position after a small change in time using the new velocity:
$$ x_{i+1} = x_i +v_{i+1}\Delta t $$

```python
# t0 is the time it starts at 
# R0 is the initial v and x values
# k and q are the constants in the force function
def EulerCromer(f, dt, t0, R0, k, q, m, imax):
    output = np.empty((imax, 3))
    i = 0
    ti = t0
    Ri = R0
    while i < imax:
        F = f(Ri[1], k, q, m)
        Ri[0] = Ri[0] + dt*F
        Ri[1] = Ri[1] + dt*Ri[0]
        ti+=dt
        output[i, 0] = ti
        output[i, 1] = Ri[0]
        output[i, 2] = Ri[1]
        i+=1
    return output
```

```python
def Force(x, k, q, m):
    F = (k*x-4*q*x**3)/m
    return F

def Kin(v, m):
    K = (m*v**2)/2
    return K

def Pot(x, k, q):
    U = -(k*x**2)/2+q*x**4
    return U
```

Firstly the system of equations will be evaluated for the k = 1N/m and q = 0.1 N/m$^3$. The initial velocity will be zero as the object is released from rest. This will then be evaluated at multiple different initial displacements:

    (a) x = 0.001m
    (b) x = 0.01m
    (c) x = 0.1m
    (d) x = 1m
    (e) x = 10m
    
    
After a solution for each is found a plot of the potential and kinetic energies will be plotted with time.

```python
def EnergyplotEC(f,dt,t,v,k,q,m,imax):
    for i in range(0,5):
        output = np.empty((imax,5))
        j = 0
        tj = t
        vj = v
        xj = 10**(-3+i)
        while j < imax:
            F = f(xj, k, q, m)
            vj = vj + dt*F
            xj = xj + dt*vj
            tj+=dt
            output[j,0] = tj
            output[j,1] = vj
            output[j,2] = xj
            output[j,3] = (m*(vj)**2)/2
            output[j,4] = -(k*(xj)**2)+q*((xj)**4)
            j+=1
        K = output[:,3]
        U = output[:,4]
        plt.plot(t, U, 'r-')
        plt.plot(t, K, 'b-')
        plt.set_xlabel('Time (t)')
        plt.set_ylabel('Energy (J)')
    plt.show()
```

```python
EnergyplotEC(Force, 0.1, 0, 0, 1, 0.1, 1, 200)
```

```python
# Solutions for each set of initial positions
sola = EulerCromer(Force, 0.1, 0, [0,0.001], 1, 0.1, 1, 100)
solb = EulerCromer(Force, 0.1, 0, [0,0.01], 1, 0.1, 1, 100)
solc = EulerCromer(Force, 0.1, 0, [0,0.1], 1, 0.1, 1, 100)
sold = EulerCromer(Force, 0.1, 0, [0,1], 1, 0.1, 1, 100)
sole = EulerCromer(Force, 0.1, 0, [0,10], 1, 0.1, 1, 100)

# Data to be plotted for each set of solutions (time,kinetic energy and potential energy)
ta = sola[:,0]
Ka = Kin(sola[:,1], 1)
Ua = Pot(sola[:,2], 1, 0.1)
tb = solb[:,0]
Kb = Kin(solb[:,1], 1)
Ub = Pot(solb[:,2], 1, 0.1)
tc = solc[:,0]
Kc = Kin(solc[:,1], 1)
Uc = Pot(solc[:,2], 1, 0.1)
td = sold[:,0]
Kd = Kin(sold[:,1], 1)
Ud = Pot(sold[:,2], 1, 0.1)
te = sole[:,0]
Ke = Kin(sole[:,1], 1)
Ue = Pot(sole[:,2], 1, 0.1)



fig, axs = plt.subplots(5, figsize = (10,20))

plt.subplots_adjust(top=0.95,hspace = 0.3)

plt.suptitle('Plot of kinetic and potential energies')
axs[0].plot(ta,Ua,'r-')
axs[0].plot(ta,Ka,'b-')
axs[0].set_title('x\u2080 = 0.001m')
axs[0].set_ylabel('Energy (J)')
axs[0].set_xlabel('Time (t)')

axs[1].plot(tb,Ub,'r-')
axs[1].plot(tb,Kb,'b-')
axs[1].set_title('x\u2080 = 0.01m')
axs[1].set_ylabel('Energy (J)')
axs[1].set_xlabel('Time (t)')

axs[2].plot(tc,Uc,'r-')
axs[2].plot(tc,Kc,'b-')
axs[2].set_title('x\u2080 = 0.1m')
axs[2].set_ylabel('Energy (J)')
axs[2].set_xlabel('Time (t)')

axs[3].plot(td,Ud,'r-')
axs[3].plot(td,Kd,'b-')
axs[3].set_title('x\u2080 = 1m')
axs[3].set_ylabel('Energy (J)')
axs[3].set_xlabel('Time (t)')

axs[4].plot(te,Ue,'r-')
axs[4].plot(te,Ke,'b-')
axs[4].set_title('x\u2080 = 10m')
axs[4].set_ylabel('Energy (J)')
axs[4].set_xlabel('Time (t)');
```

$$F(x) = -kx$$
$$U(x) = \frac{kx^2}{2}$$

#### Rungga-Kutta Method
Another method for solving the this system of ODEs is utilising the Rungga-Kutta method.
This has the following steps:
    
    1. Define k1 and k2.
$$ k_1 = \frac{F(x_i)}{m} \Delta t $$
$$ k_2 = \frac{F(x_i+dx)}{m} \Delta t $$
    2. Calculate the next velocity, v, and displacement using these values for k1 and k2.
$$ v_{i+1} = v_i+\frac{1}{2}(k_1+k_2)$$
$$ x_{i+1} = x_i+ v_{i+1} \Delta t $$


```python
def RK2(f, dt, t0, R0, k, q, m, imax):
    output = np.empty((imax, 3))
    i = 0
    ti = t0
    Ri = R0
    while i < imax:
        k1 = dt*f(Ri[1], k, q, m)
        k2 = dt*f(Ri[1]+k1, k, q, m)
        Ri[0] = Ri[0] + 0.5*(k1 + k2)
        Ri[1] = Ri[1] + dt*Ri[0]
        ti += dt
        output[i, 0] = ti
        output[i, 1] = Ri[0]
        output[i, 2] = Ri[1]
        i += 1
    return output
```

```python
# Solutions for each set of initial positions
sola2 = RK2(Force, 0.1, 0, [0,0.001], 1, 0.1, 1, 100)
solb2 = RK2(Force, 0.1, 0, [0,0.01], 1, 0.1, 1, 100)
solc2 = RK2(Force, 0.1, 0, [0,0.1], 1, 0.1, 1, 100)
sold2 = RK2(Force, 0.1, 0, [0,1], 1, 0.1, 1, 100)

# Data to be plotted for each set of solutions (time,kinetic energy and potential energy)
ta2 = sola2[:,0]
Ka2 = Kin(sola2[:,1], 1)
Ua2 = Pot(sola2[:,2], 1, 0.1)
tb2 = solb2[:,0]
Kb2 = Kin(solb2[:,1], 1)
Ub2 = Pot(solb2[:,2], 1, 0.1)
tc2 = solc2[:,0]
Kc2 = Kin(solc2[:,1], 1)
Uc2 = Pot(solc2[:,2], 1, 0.1)
td2 = sold2[:,0]
Kd2 = Kin(sold2[:,1], 1)
Ud2 = Pot(sold2[:,2], 1, 0.1)



fig, ax1 = plt.subplots(4, figsize = (10,16))

plt.subplots_adjust(top=0.95,hspace = 0.3)

plt.suptitle('Plot of kinetic and potential energies')
ax1[0].plot(ta2,Ua2,'r-')
ax1[0].plot(ta2,Ka2,'b-')
ax1[0].set_title('x\u2080 = 0.001m')
ax1[0].set_ylabel('Energy (J)')
ax1[0].set_xlabel('Time (t)')

ax1[1].plot(tb2,Ub2,'r-')
ax1[1].plot(tb2,Kb2,'b-')
ax1[1].set_title('x\u2080 = 0.01m')
ax1[1].set_ylabel('Energy (J)')
ax1[1].set_xlabel('Time (t)')

ax1[2].plot(tc2,Uc2,'r-')
ax1[2].plot(tc2,Kc2,'b-')
ax1[2].set_title('x\u2080 = 0.1m')
ax1[2].set_ylabel('Energy (J)')
ax1[2].set_xlabel('Time (t)')

ax1[3].plot(td2,Ud2,'r-')
ax1[3].plot(td2,Kd2,'b-')
ax1[3].set_title('x\u2080 = 1m')
ax1[3].set_ylabel('Energy (J)')
ax1[3].set_xlabel('Time (t)')
```

The Rungga-Kutta method was unable to evaluate for a displacement of 10m as it became extremely large in calculation with the cubed position term. The potential peaks overlap the kinetic energy peaks for the Rungga-Kutta method unlike the Euler-Cromer method.
