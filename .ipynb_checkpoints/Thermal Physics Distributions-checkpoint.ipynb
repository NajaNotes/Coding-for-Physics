```python
def Gaussian(x,m,sigma2):
    gaussian = (1/np.sqrt(sigma2*2*np.pi))*np.exp(-(x-m)**2/(2*sigma2))
    return gaussian
```

## Physics Distributions
- Boltzmann
- Bose-Einstein 
- Fermi-Dirac 

```python
def Boltzmann(E):
    return np.exp(-E)

def Bose(E):
    return 1/(np.exp(E)-1)

def Fermi(E):
    return 1/(1+np.exp(E))
```

```python
y = np.linspace(0,3,200)
z = np.linspace(0.025,3,200)
B = Boltzmann(y)
BE = Bose(z)
FD = Fermi(y)
plt.title('Physics Distributions')
plt.plot(y,B,'r-',label='Boltzmann Distribution')
plt.plot(y,BE,'g-',label='Bose-Einstein Distribution')
plt.plot(y,FD,'b-',label='Fermi-Dirac Distribution')
plt.ylim(0,6)
plt.ylabel('Number of particles in each state')
plt.xlim(0,3)
plt.xlabel('(E-\u03BC)/kT')
plt.legend();
```

### Boltzmann Distribution 
This is also known as the Maxwell-Boltzmann distribution and is very important in thermodynamics and is given by:

$$\bar{n} = e^{-\frac{\epsilon-\mu}{k_BT}}$$


### Bose-Einstein Distribution

The average number of particles in a state with energy $\epsilon$ is:
$$\bar{n} = \frac{1}{e^{\frac{\epsilon-\mu}{k_BT}}-1}$$

This equation arises from the fact that any number of bosons can exist in a state at a given energy.

### Fermi-Dirac Distribution

The average number of fermions in a state with energy $\epsilon$ is:

$$\bar{n} = \frac{1}{e^{\frac{\epsilon-\mu}{k_BT}}+1}$$

This equation arises from the Pauli Exclusion Principle. The Pauli Exclusion Principle ensures that a state at a given energy either has 1 or no fermions as no 2 fermions can have the same energy and quantum state. 

### Blackbody Spectrum
##### - Rayleigh-Jeans Law

$$I(\lambda,T) = \frac{2\pi ckT}{\lambda^4}$$

##### - Planck's Law

$$I(\lambda,T) = \frac{2\pi hc^2}{{\lambda}^5(e^{(hc/\lambda kT)}-1)}$$

```python
def Planck(l):
    return 1/((l**5)*(np.exp(1/l)-1))


def RayJeans(l):
    return l**(-4)

a = np.linspace(0.05,1,100)
b = np.linspace(0.3,1,100)
RJ = RayJeans(b)
Pl = Planck(a)

plt.title('Blackbody Spectrum')
plt.plot(b,RJ,'r-',label='Rayleigh-Jeans Law')
plt.plot(a,Pl,'g-',label='Plancks Law')
plt.xlabel('Wavelength \u03BB (m)')
plt.xlim(0,1)
plt.ylabel('Flux')
plt.legend();
```

## Diffusion in air
As a particle travels through the air it will experience elastic collisions between each particle and will travel a distance given by the formula:
$$ d = \lambda\sqrt N$$

```python
def diffusion(l,N):
    return l*(N)**(0.5)
```

```python
# collision is maximum number of collisions
# path is number of paths 
def Steptravel(collision,path):
    # array for the mean travel after N collisions 
    m = np.zeros((3,collision))
    # for each number of collisions
    for i in range(0,collision):
        paths = np.zeros(path)
        for j in range(0,path):
            position = np.zeros(2)
            for k in range(1,i+2):
                x = np.random.uniform(0,2*np.pi)
                position+=np.array([np.cos(x),np.sin(x)])
            paths[j] = np.sqrt(np.dot(position,position))
        m[0,i] = np.sum(paths)/(path)   
        m[1,i] = i+1
        m[2,i] = np.sqrt(np.dot((paths-m[0,i]),(paths-m[0,i]))/(path-1))
    return m 
x3 = Steptravel(500,1000) 
```

```python
plt.figure(figsize=(12,7))

x4 = np.linspace(0,500,500)

plt.title('Mean distance travelled between collisions')
plt.xlabel('Number of collisions')
plt.ylabel('Mean distance travelled (m)')
plt.plot(x3[1],x3[0],'b-')
plt.plot(x4,diffusion(1,x4));
```

## Potential in Alpha Decay

The barrier that an $\alpha$-particle must escape is dependent on the coulomb force potential and the total angular momentum of the nucleus:

$$U = \frac{Zze^2}{4\pi \epsilon_or}+\frac{l(l+1)\hbar^2}{2Mr^2}$$

```python
def AlphaPot(r,l,M,Z):
    return Z*2/(4*np.pi*(5.5263e11)*r)+l*(l+1)*(1.97328e-13)**2/(2*M*(r*1e-2)**2)

plt.figure(figsize=(12,7))

r1 = np.linspace(7.499e-13,5e-12,100)
plt.plot(r1,AlphaPot(r1,0,3727.379,94),label='l=0')
plt.plot(r1,AlphaPot(r1,1,3727.379,94),label='l=1')
plt.plot(r1,AlphaPot(r1,2,3727.379,94),label='l=2')
plt.plot(r1,AlphaPot(r1,10,3727.379,94),label='l=10')
plt.legend();

print('barrier(l=0) =',AlphaPot(7.499e-13,0,3727.379,94))
print('barrier(l=1) =',AlphaPot(7.499e-13,1,3727.379,94))
print('barrier(l=2) =',AlphaPot(7.499e-13,2,3727.379,94))
print('barrier(l=10) =',AlphaPot(7.499e-13,10,3727.379,94))
```
